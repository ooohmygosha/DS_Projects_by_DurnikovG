# Восстановление золота из руды

## Введние

**Цель проекта:**

Смоделировать процесс восстановления золота из золотосодержащей руды с помощью модели машинного обучения.

**Задача:**

* Построить модель, способную предсказать коэффициент восстановления золота из золотосодержащей руды, используя данные с параметрами добычи и очистки. 

**Целевые метрики:**
* **sMAPE** -  sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
Рассчитывается по следющей формуле:
![](https://pictures.s3.yandex.net/resources/smape_1576238825.jpg)

**Стартовый набор данных:**

В нашем распоряжении данные с параметрами добычи и очистки. Нужно построить модель для задачи регересии, которая предскажет нам коэффициент восстановления золота из золотосодержащей руды.

Коэффициент рассчитывается следующим образом:
![](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)

где:
* **C** — доля золота в концентрате после флотации/очистки;
* **F** — доля золота в сырье/концентрате до флотации/очистки;
* **T** — доля золота в отвальных хвостах после флотации/очистки.


Данные находятся в трёх файлах и уже предварительно разбиты на обучающую и тестовую выборки:
* gold_recovery_train.csv — обучающая выборка
* gold_recovery_test.csv — тестовая выборка
* gold_recovery_full.csv — исходные данные


Описание данных:

**Технологический процесс:**
* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

**Параметры этапов:**
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи


**Наименование признаков:**
* [этап].[тип_параметра].[название_параметра]  
**Пример:** rougher.input.feed_ag

**Возможные значения для блока [этап]:**
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики


**Возможные значения для блока [тип_параметра]:**
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

**Нужно спрогнозировать сразу две величины:**
1. Эффективность обогащения чернового концентрата **rougher.output.recovery**;
2. Эффективность обогащения финального концентрата **final.output.recovery**.

**Итоговая метрика складывается из двух величин по следующей формуле:**
![](https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg)

### Несколько основных функций и констант

```python
RANDOM_STATE = 101
```

Функция для показа основной информации о датафрейме.

```python
def meet(df):
    display(df.head(10), df.tail(10))
    df.info()
```

Функция для вывода кол-ва пропущенных значений в датафрейме.

```python
def perc_miss(df):
    display(df.isna().mean()[df.isna().mean() > 0].sort_values().reset_index().rename(
    columns={'index': 'feature',
             0: 'percentage of missing values'}
))
```

Функция для вычисления sMAPE для одного таргета. Принимает на вход фактические и предсказанные значения. Возвращает sMAPE.

```python
def smape(y_test, y_pred, **kwargs):
    return 1/len(y_test) * np.sum(2 * np.abs(y_pred-y_test) / (np.abs(y_test) + np.abs(y_pred))*100)
```

С помощью функции make_scorer сделаем свою метрику оценки, для передачи в параметры кросс-валидации и обучения.

```python
smape_score = make_scorer(smape, greater_is_better=False)
```

Функция для расчета итогового значения взвешенного sMAPE. sMAPE для флотации  с весом 0,25 + sMAPE для финальной стадии с весом 0,75.

```python
def total_smape(score_rough, score_final):
    return score_rough*0.25 + score_final*0.75
```

### Подбор гиперпараметров для модели LGBMRegressor  с помощью Optuna

Для подбора гиперпарметров нужна кастоизированная функция метрики.

```python
def lgbm_smape(y_true, y_pred):
    smape_val = smape(y_true, y_pred)
    return 'SMAPE', smape_val, False
```

Функция для подсчета скора на модели, с найдены оптимизированными параметрами.

```python
def model_test(params, name):
    model = LGBMRegressor(**params)

    model.fit(X_train,
              y_train['rougher.output.recovery'],
              eval_metric=lambda y_true, y_pred: [lgbm_smape(y_true, y_pred)],
              eval_set=[(X_test,y_test['rougher.output.recovery'])],
              early_stopping_rounds=100,
              verbose=500)
    rough_preds = model.predict(X_test)
    rough_test_smape = smape(y_test['rougher.output.recovery'], rough_preds)


    model.fit(X_train,
              y_train['final.output.recovery'],
              eval_metric=lambda y_true, y_pred: [lgbm_smape(y_true, y_pred)],
              eval_set=[(X_test,y_test['final.output.recovery'])],
              early_stopping_rounds=100,
              verbose=500)
    final_preds = model.predict(X_test)
    final_test_smape = smape(y_test['final.output.recovery'], final_preds)

    total_smape(rough_test_smape, final_test_smape)

    MODEL_NAME.append(name)
    ROUGH_SMAPE.append(rough_test_smape)
    FINAL_SMAPE.append(final_test_smape)
    TOTAL_SMAPE.append(total_smape(rough_test_smape, final_test_smape)) 
```

Для начала проверим на модели без тюнинга гиперпараметров.

```python
display_smape()
```

Получили скор 9,73. Уже лучше чем бйзлайн, но можем ли мы дать лучший перфоманс с этой моделью?

* Описываем функцию для подбора гиперпараметров. 
* Укзаываем в ней все пространство подбора для основных гиперпараметров.

```python
def objective(trial,data=X_train,target=y_train['final.output.recovery']):
    
    train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.2,random_state=RANDOM_STATE)
    param = {
        'random_state': RANDOM_STATE,
        'n_estimators': 20000,
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-3, 10.0),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-3, 10.0),
        'colsample_bytree': trial.suggest_categorical('colsample_bytree', [0.3,0.4,0.5,0.6,0.7,0.8,0.9, 1.0]),
        'subsample': trial.suggest_categorical('subsample', [0.4,0.5,0.6,0.7,0.8,1.0]),
        'learning_rate': trial.suggest_categorical('learning_rate', [0.006,0.008,0.01,0.014,0.017,0.02]),
        'max_depth': trial.suggest_categorical('max_depth', [10,20,100]),
        'num_leaves' : trial.suggest_int('num_leaves', 1, 1000),
        'min_child_samples': trial.suggest_int('min_child_samples', 1, 300),
        'cat_smooth' : trial.suggest_int('min_data_per_groups', 1, 100)
    }
    model = LGBMRegressor(**param)  
    
    model.fit(train_x,
              train_y,
              eval_metric=lambda y_true, y_pred: [lgbm_smape(y_true, y_pred)],
              eval_set=(test_x,test_y),
              early_stopping_rounds=100,
              verbose=False)
    
    preds = model.predict(test_x)
    
    smape_score = smape(test_y, preds)
    
    return smape_score
```

Запускаем процесс обучения Optuna из 50 попыток.

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

```python
params = study.best_params   
params['random_state'] = RANDOM_STATE
params['n_estimators'] = 20000 
params['cat_smooth'] = params.pop('min_data_per_groups')
```

```python
model_test(params, 'Test LGBM Optuna')
```

### Подбор гиперпараметров для LGBM с помощью Hyperopt

Формируем датасеты, описываем функцию для проведения оптимизации.

```python
lgtrain = lgbm.Dataset(X_train, label=y_train['final.output.recovery'])
lgtest = lgbm.Dataset(X_test, label=y_test['final.output.recovery'])

def evaluate_metric(params):
    
    model_lgb = lgbm.train(params, lgtrain, 500, 
                          valid_sets=[lgtrain, lgtest], early_stopping_rounds=100, 
                          verbose_eval=500)

    pred = model_lgb.predict(X_test)

    score = smape(y_test['final.output.recovery'],pred)
    
    print(score)
 
    return {
        'loss': score,
        'status': STATUS_OK,
        'stats_running': STATUS_RUNNING
    }
```

Запускаем процес оптимизации, тоже 50 повторений.

```python
trials = Trials()

MAX_EVALS = 50

best_vals = fmin(evaluate_metric, space=hyper_space, verbose=1,
                 algo=tpe.suggest, max_evals=MAX_EVALS, trials=trials)

best_params = space_eval(hyper_space, best_vals)
```
