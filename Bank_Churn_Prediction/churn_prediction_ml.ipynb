# Предсказание оттока клиентов

## Введение

**Цель проекта:**

Проведя анализ исторических данных о поведении клиентов и расторжении договоров с "NDA Банком" предсказать отток для текущих клиентов.

**Задача:**

* Построить модель, способную спрогнозировать уйдёт клиент из банка в ближайшее время или нет. 

**Целевые метрики:**
* **F1-sсore** - гармоническое среднее precision и recall. Показывает то, как точно мы определяем "отточных" клитентов, при этом не даем ложно-отрицательных и ложно-положительных рекомендаций. Проще говоря - максимально точечно расходуем ресурсы по пресечению оттока "отточных" клиентов. При этом не расходуя ресурсы на тех кто "отточными" клиентами не являются. Условием заказчика явлется довести метрику до **0.59**. 
* **ROC-AUC-score** - площадь, ограниченная ROC-кривой и осью доли ложных положительных классификаций. Чем выше показатель ROC-AUC-score, тем качественнее маша модель в целом определяет кажого из клиентов (обоих классов).

**Стартовый набор данных:**

В нашем распоряжении исторические данные о поведении клиентов. Нужно построить модель для задачи классификации, которая предскажет нам, кто из текущих клиентов склонен у уходу. 

Каждый объект в наборе данных — это один уникальный клиент.


Признаки:  
* RowNumber — индекс строки в данных
* CustomerId — уникальный идентификатор клиента
* Surname — фамилия
* CreditScore — кредитный рейтинг
* Geography — страна проживания
* Gender — пол
* Age — возраст
* Tenure — сколько лет человек является клиентом банка
* Balance — баланс на счёте
* NumOfProducts — количество продуктов банка, используемых клиентом
* HasCrCard — наличие кредитной карты
* IsActiveMember — активность клиента
* EstimatedSalary — предполагаемая зарплата

Целевой признак:
* Exited — факт ухода клиента

1. Подгружаем либы, опишем все функции, которые пригодятся в процессе анализа.

8. Каждую из выбранных моделей проверим на тестовых данных.
9. Итоговый вывод.

Функция для посторения хитмапа по корреляции между фичами.

```python
def heatmap_corr(df, title):
    fig, ax = plt.subplots(figsize=(16, 10))
    corr = df.corr()
    return sns.heatmap(corr, cmap="rocket_r", annot=True,
                       linewidths=.5, ax=ax).set_title(title)
```

Функция для посторения матрицы несоотвествий для предикта.

```python
def plot_confusionmatrix(y_train_pred, y_train, dom):
    classes = ['Not Exited', 'Exited']
    print(f'{dom} Confusion matrix')
    cf = confusion_matrix(y_train_pred, y_train)
    sns.heatmap(cf/np.sum(cf), annot=True, yticklabels=classes,
                xticklabels=classes, cmap='rocket_r', fmt='.2%')
    plt.tight_layout()
    plt.show()
```

Функция для оверсемплинга минорного класса.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=RANDOM_STATE)
    
    return features_upsampled, target_upsampled
```

Функция для вывода информации о получившейся моедли.
* Обучает модель по обучающим данным. 
* Сообщает F1-score, ROC-AUC score для обучющей и валидационной выборок, а так же строит материцу несоотвествий.

```python
def model_report(name, model, x_train, y_train):
    
    
    if name == 'CatBoostClassifier':
        model.fit(x_train, y_train, eval_set=(x_val, y_val))
    else:
        model.fit(x_train, y_train)
    
    y_pred_train = model.predict(x_train)
    y_pred_train_proba = model.predict_proba(x_train)[:,1]
    y_pred = model.predict(x_val)
    y_pred_proba = model.predict_proba(x_val)[:,1]
    
    
    print(f'{name} Train f1 score {f1_score(y_train, y_pred_train)}')
    print(f'{name} Train roc auc score {roc_auc_score(y_train, y_pred_train_proba)}')
    print()
    
    plot_confusionmatrix(y_pred_train, y_train, dom=f'{name} Train')
    print()
    
    print(f'{name} Validation f1 score {f1_score(y_val, y_pred)}')
    print(f'{name} Validation roc auc score {roc_auc_score(y_val, y_pred_proba)}')
    print()
    
    plot_confusionmatrix(y_pred, y_val, dom=f'{name}  Valid')
    
```

### Отладка методов оверсэмплинга.

#### SMOTE + Undersampling

* Алгоритм SMOTE строит новые точки на основе образованных связей между имеющимися значениями. Таким образом происходит оверсэмплинг. В комбинации с ним используем андерсэмплинг мажорного класса до соотношения маж./мин. = 2/1.


* Переберем в цикле кол-во "соседей" и sampling_strategy для метода SMOTE. В качестве модели выбреем "недообученный" и быстрый  LGBMClassifier. 


* Используя кросс валидацию с RepeatedStratifiedKFold почитаем средний F1-scorе.


* Исходя из лучшего F1-score выберем настройки метода. Оверсемплированные обучающие выборки передадим в качетсве аргументов всем исследуемым функциям. 

```python
k_values = [1, 2, 3, 4, 5, 6, 7]
for k in k_values:
    model = LGBMClassifier(metric='auc', random_state=RANDOM_STATE)
    for i in range(3,6,1):
        over = SMOTE(sampling_strategy=i*0.1, k_neighbors=k, random_state=RANDOM_STATE)
        under = RandomUnderSampler(sampling_strategy=0.5, random_state=RANDOM_STATE)
        steps = [('over', over), ('under', under), ('model', model)]
        pipeline = Pipeline(steps=steps)
        cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=RANDOM_STATE)
        model_scores = cross_val_score(pipeline, x_train, y_train, scoring='f1', cv=cv, n_jobs=-1)
        model_score = np.mean(model_scores)
        
        print('> k=%d,i=%.3f Mean f1: %.3f' % (k,i*0.1, model_score))
```

Максимальные значения у 
* **k=5,i=0.300 / Mean f1: 0.618**

1. Для каждой модели проведем кросс-валдиацию и посчитаем F1-score, ROC-AUC score.
2. Затем обучим исследуемые модели и сделаем предикт. Посчитаем F1-score, ROC-AUC score для валидационной выборки.

#### ADASYN

ADASYN улучшенная версия Smote. Он добавляет случайные небольшие значения к точкам, делая их более реалистичными.

Аналогичным путем выберем показатели для него. 

```python
k_values = [1, 2, 3, 4, 5, 6, 7]
for k in k_values:
    model = LGBMClassifier(metric='auc', random_state=RANDOM_STATE)
    for i in range(3,6,1):
        over = ADASYN(sampling_strategy=i*0.1, n_neighbors=k, random_state=RANDOM_STATE)
        under = RandomUnderSampler(sampling_strategy=0.5, random_state=RANDOM_STATE)
        steps = [('over', over), ('under', under), ('model', model)]
        pipeline = Pipeline(steps=steps)
        cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=RANDOM_STATE)
        model_scores = cross_val_score(pipeline, x_train, y_train, scoring='f1', cv=cv, n_jobs=-1)
        model_score = np.mean(model_scores)
        
        print('> k=%d,i=%.3f Mean f1: %.3f' % (k,i*0.1, model_score))
```

Максимальные значения у

#### BY-HAND Upsampling

Третий метод оверсэмплинга - все значения минорного класса берем в цикле для каждой модели по нескольку раз - от 1 до 4. Перемешиваем.  
Таким образом проверям соотношения 1/4, 2/4, 3/4, 4/4.

```python
upsampling_coef = range(1, int(np.ceil((1-y.mean())/ y.mean())+1))
```

```python
f1_scores = []
ra_scores = []
for model, name in zip(best_models,best_models_names):
    predictions = model.predict(x_test)
    f1 = f1_score(y_test, predictions)
    roc_auc = roc_auc_score(y_test, model.predict_proba(x_test)[:,1])
    f1_scores.append(f1)
    ra_scores.append(roc_auc)
    print(f'{name} Test f-1 score {f1}')
    print(f'{name} Test roc_auc score  {roc_auc}')
    print()
    plot_confusionmatrix(predictions, y_test, dom=f'{name} Test')
```

```python
final_scores_df = pd.DataFrame({'name': best_models_names,
                                'best_test_f1_score': f1_scores,
                                'best_test_roc_auc_score': ra_scores})
final_scores_df.sort_values(by='best_test_f1_score', ascending=False, inplace=True)
final_scores_df
```
