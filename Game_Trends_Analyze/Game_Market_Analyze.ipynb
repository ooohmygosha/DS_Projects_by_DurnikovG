### Функции которые пригодятся в процессе анализа

Функция для получения основной информации о датасете.

Функция для получения доли пропущенных значений в датасете.

```python
def perc_miss(df):
    return df.isna().mean()[df.isna().mean() > 0].sort_values().reset_index().rename(
    columns={'index': 'feature',
             0: 'percentage of missing values'}
)
```

Функция для построения хитмапа корреляции между признаками в датасете.

```python
def heatmap_corr(df, title):
    fig, ax = plt.subplots(figsize=(16, 10))
    corr = df.corr()
    return sb.heatmap(corr, cmap="rocket_r", annot=True, linewidths=.5, ax=ax).set_title(title)
```

Функция для группировки по столбцу и  расчета значений.

```python
def group_feature(df, feature):
    bp = df.groupby(feature).agg(
    {'name':'count', 'global_sales': 'sum'})
    bp = bp.reset_index()
    bp['sales_per_game'] = bp.global_sales / bp.name
    bp = bp.sort_values(by='sales_per_game', ascending=False).reset_index(drop=True)
    return bp
    
```

Функция для построения диаграммы рассеивания по двум признакам.

```python
def scatplot(data, x, y, hue, title):
    plt.figure(figsize=(20, 12))
    sb.scatterplot(data=data,
               y=y,
               x=x,
               hue=hue,
               palette="rocket_r").set_title(title)                      
```

Функция для построения диаграммы рассеивания по двум признакам и регрессионной прямой.

```python
def regplot(data, x, y, title):
    plt.figure(figsize=(20, 12))
    sb.regplot(data=data,
               y=y,
               x=x).set_title(title)  
```

Функция для построения столбчатой диаграммы для опредления популярных фичей относительно региона.

```python
def popular(df,x,y):
    bar = df.groupby(x)[y].sum().reset_index().sort_values(by=y, ascending=False).head(5)
    title = 'Total sales for ' +  y[:2].upper() + ' region for each ' + x
    plt.figure(figsize=(16, 9))
    sb.barplot(x=x,y=y,
             data=bar,
             palette="rocket_r").set_title(title)
```

#### Посмотрим сколько игры выпускалось за каждый год.

```python
def group_feature(df, feature):
    bp = df.groupby(feature).agg(
    {'name':'count', 'global_sales': 'sum'})
    bp = bp.reset_index()
    bp['sales_per_game'] = bp.global_sales / bp.name
    bp = bp.sort_values(by='sales_per_game', ascending=False).reset_index(drop=True)
    return bp
    
```

```python
df_year = group_feature(df, 'year_of_release')
```

```python
plt.subplots(figsize=(20,12))
sb.countplot(df['year_of_release'], palette="rocket_r").set_title('Releases per year');
```

## Проверка статистических гипотез

Для принятия решения о ходе рекламной компании проверим две гипотезы:

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action и Sports разные.


Пороговое значение alpha задаем рекомендованное для эконометрики 0,05.

Ниже приведены комментарии:  
* как формулировали нулевую и альтернативную гипотезы. 
* какой критерий использовали для проверки гипотез.

Сформируем датафрейм для оценки гипотез.

```python
meet(df)
```

```python
user_score_columns = ['name', 'platform', 'genre', 'user_score'] 
df_userscore = df[user_score_columns]
df_userscore = df_userscore[df_userscore.user_score >= 0]
meet(df_userscore)
```

Датафрейм для анализа из 1270 строк имющие оценку игроков.

### Проверка 1. Одинаковы ли средние пользовательские рейтинги платформ Xbox One и PC?

```python
df_userscore.query('platform == "xone"').user_score.describe()
```

```python
df_userscore.query('platform == "pc"').user_score.describe()
```

**Дано:**  
Два множества данных: Игры для Xbox One и PC.

**Формулируем гипотезы:**
* H0: Средние для двух выборок равны.  
* Н1: Средние для двух выборок не равны.

Сопоставимое число записей для каждой из платформ. 

Видим, что средние для данных выборок различаются минимально. Нужно проверить яляется ли это раличие статистически значимым. Если выясним, что это не так (принимаем Н0) - значит оценки игроков сопоставимы для обоих платформ.

Диспресии для выборок отличаются, поэтому как аргумент передадим информацию о различии дисперсий для функции вычисления p-value. 

```python
pc_df = df_userscore.query('platform == "pc"').user_score
xone_df = df_userscore.query('platform == "xone"').user_score

alpha = 0.05 

results = st.ttest_ind(pc_df, xone_df, equal_var=False)

print('p-значение:{:.10}:'.format(results.pvalue)) 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не отвергаем Н0.   
**Значит различие статистически незначимо, и игроки рейтнгуют игры на PC и Xbox One одинаково.**

### Проверка 2. Разные ли средние пользовательские рейтинги жанров Action и Sports?

```python
df_userscore.query('genre == "action"').user_score.describe()
```

```python
df_userscore.query('genre == "sports"').user_score.describe()
```

**Дано:**  
Два множества данных: Игры жанров Action и Sports.

**Формулируем гипотезы:**
* H0: Средние для двух выборок равны.  
* Н1: Средние для двух выборок не равны.

Игр в жанре экшен больше в 3 раза.

Видим, что средние для данных выборок различаются достаточно серьезно:
* 6,86 для Action
* 5.46 для Sports

Нужно проверить яляется ли это раличие статистически значимым. Если выясним, что это так (отвергаем Н0) - значит оценки игроков жанра Action выше оценок жанра Sports


Диспресии для выборок отличаются, поэтому как аргумент передадим информацию о различии дисперсий для функции вычисления p-value. 

```python
acrion_df = df_userscore.query('genre == "action"').user_score
sports_df = df_userscore.query('genre == "sports"').user_score
alpha = 0.05 

results = st.ttest_ind(acrion_df, sports_df, equal_var=False)

print('p-значение:{:.10}:'.format(results.pvalue)) 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Отвергаем Н0.   
**Значит различие статистически значимо, и игроки рейтнгуют игры в жанре Action выше игр жанра Sports**

### Вывод

Используя t-тесты выяснили, что:
1. Статистически игроки рейтнгуют игры на PC и Xbox One одинаково. PC-гейминг в сердцах у игроков.


2. Статистически игроки рейтнгуют игры в жанре Action выше игр жанра Sports. И это несмотря на то, что средние проджажи в жанре Sports выше.

* Игры может продаваться хорошо, даже имея средние оценки от игроков. Влияние оценок игроков на игру оень ограниченное.
* Оценка критиков, наоборот - сильная фича. Магазину нужно ориентироваться на игры с хорошей прессой. Оценка ниже 80-75 % сильно снижает вероятность игре хорошо продаться, вне зависимости от платформы.

**Каким жанрам отдавать предпочтение?**
* В отношении Северной Америки и Европы одинаково необходимо концентрироваться на жанрах Shooter, Sports - в превую очередь, Action и RPG во вторую. 
* Для Японии нужно делать основную ставку  на RPG и Action.

**Каким игровым платформам отдавать предпочтение?**

* Для Европы - платформы нового поколения от Playstation. Xbox в гораздо меньшей степени.   
* Для Северной Америки - на обоеих платформах нового поколения, от Playstation и Xbox.   
* Для Японии  в первую очередь ориентрроваться на новый PS4 и на портативные платформы (3DS, PSV). 

**Какой рейтинг у самых популярных игр?**
* В Европе потребляют контент рейтинга М и Е в равной степени. Большая доля игры для детей. 
* Для Северной Америки гораздо более актуальны игры для рейтингов М и Т, следовательно тут аудитория более взрослая.
* В Японии своя система рейтинга контента в играх, в ходе данного анализа не была рассмотерна в виду отсутсвия данных.

Помимо этого были проведены стат. тесты в ходе которых еще раз убедилиссь, что оценки игроков слабо влияют на продажи. Возможно это обусловлено малой выборкой.(больше покупают - больше выборка рейтигующих - хуже оценка).  

Вывод в ходе статеста был сделан следующий:
* Статистически игроки рейтнгуют игры в жанре Action выше игр жанра Sports. И это несмотря на то, что средние проджажи в жанре Sports выше.
