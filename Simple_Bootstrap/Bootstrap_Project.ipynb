# Выбор локации для скважины

## Введение

**Цель проекта:**

* Определить регион и месторасположение в нем новых скважин для добычи нефти.

**Задача:**

* Построить модель линейной регрессии, которая поможет определить регион, где добыча принесёт наибольшую прибыль.  
Линейная регрессия является обязательным условием из-за ее интерпретируемости.


**Целевые метрики:**
* **RMSE** - метрика определения качества моедли предсказания, корень из среднеквадратичной ошибки прогнозов модели. 
* **Средняя прибыль региона** - средняя прибыль при условии бурения в топ-200 скважин по предсказаниям модели. Отдельно рассмотри ее 95%-й доверительный интервал.
* **Риск убыка** - вероятность убытка, при разработке в данном регионе. Не должна превышть 2,5%.

**Стартовый набор данных:**

Предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов.  


Данные геологоразведки трёх регионов находятся в файлах:

geo_data_0.csv  
geo_data_1.csv  
geo_data_2.csv  

Признаки:
* **id** — уникальный идентификатор скважины;
* **f0**, **f1**, **f2** — три признака точек, данные синтетические;

Целевой признак:
* **product** — объём запасов в скважине (тыс. баррелей).

**Условия задачи:**

* При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
* Бюджет на разработку скважин в регионе — 10 млрд рублей.
* При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.

**Шаги для выбора локации:**

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.


**План работы:**

1. Подгружаем либы, опишем все функции, которые пригодятся в процессе анализа.
2. Считаем файлs, познакомимся с данными. Проведем краткий EDA - посмотрим распределения и диаграммы рассеивания, поищем зависимости между фичами.
3. Подготовим данные для посторения моеделей. Разобьем данные на обучающую и валидационную выборки в соотношении 75:25. 
4. Проверим модель на кросс-валидации. Сделаем предсказания для каждого региона. Посчитаем RMSE для трэйна на кросс-валидации и валидацонной выборки.
5. Посчитаем средний предсказанный объем нефти для всех месторождений региона. 
6. Посчитаем, какой должен быть средний обхем, при текущих условиях стоимоти нефти и стоимости разработки месторождения. 
7. Проанализиреум риски и веорятную среднюю прибыль для каждого региона с помощью Bootstrap.
8. Выберем регион исходя из условий поставленной задачи.
9. Итоговый вывод.

```python
def heatmap_corr(df, title):
    fig, ax = plt.subplots(figsize=(16, 10))
    corr = df.corr()
    return sns.heatmap(corr, cmap="rocket_r", annot=True,
                       linewidths=.5, ax=ax).set_title(title)
```

## Обучение и проверка модели

Функция для построения отчета о регионе.

1. На вход принимает датафрейм.
2. Делит на обучающие и валидационную выборки.
3. Проводит кросс-валидациую
4. Обучается на трейн-данных, делает предикт на валидационные.
5. Возвращает датафрейм с правильным таргетом и предсказнным. 
6. Выводит на экран отчет о регионе.

```python
def region_report(df):
    
    x = df.drop(['id', 'product'], axis=1)
    y = df['product']
    
    bins = np.linspace(0, len(y), 50)
    y_binned = np.digitize(y, bins)
    
    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y_binned)
    
    rmse_baseline = mean_squared_error(y_val, [y_train.mean()]*len(y_val), squared=False)
    
    scalar = StandardScaler()
    model = LinearRegression()

    pipeline = Pipeline([('transformer', scalar), ('estimator', model)])

    cv = RepeatedKFold(n_splits=5, n_repeats=5,random_state=RANDOM_STATE)
    scores = cross_val_score(pipeline, x_train, y_train, cv=cv,scoring='neg_mean_squared_error')
    mean_score = scores.mean()
    rmse_cv = np.sqrt(-mean_score)
       
    x_train = scalar.fit_transform(x_train)
    x_val = scalar.transform(x_val)
   
    model.fit(x_train, y_train)
    y_pred = model.predict(x_val)
    rmse_test = mean_squared_error(y_val, y_pred, squared=False)
    diff = pd.DataFrame({'product_true':y_val, 'product_pred':y_pred})
    
    print(f'RMSE of the baseline model for {namestr(df, globals())[0]} region: {rmse_baseline:.2f}')
    print(f'Cross-validation RMSE of the model for {namestr(df, globals())[0]} region: {rmse_cv:.2f}')
    print(f'Valid RMSE of the model for {namestr(df, globals())[0]} region: {rmse_test:.2f}')
    print()
    print(f'Forecasting average stock for {namestr(df, globals())[0]} region: {diff.product_pred.mean():.2f} thousand barrels')
    
    return diff
```

```python
diff_df_0 = region_report(df_0)
```

```python
diff_df_2 = region_report(df_2)
```

**Вывод:**
* Лучший предсказнный средний запас для местороджения - регион "2", 95,05 тыс. баррелей. Но самый низкий RMSE, действительно близкий к бэйзлайну. 
* На втором месте по предсказанию - регион "0", RMSE уже ощутимо лучше бэйзлайна. Пока выглядит саммым перспективным.
* Регион "1" со своей разметкой показал RMSE меньше 1, что конечно указывает на то, что серьезно это прогноз рассматривать не нужно.


## Подготовка к расчёту прибыли

Вернемся к задаче:

При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
Бюджет на разработку скважин в регионе — 10 млрд рублей.

При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.

Таким образом нужно посчитать, какой минимальный средний объем должен быть у месторождений, чтобы хотябы не было убытка.

```python
BUDGET = 10_000_000_000
PRICE = 450_000
SEARCH_WELLS = 500
TOP_WELLS = 200
BUDGET_PER_WELL = BUDGET / TOP_WELLS
```

Бюджет-на-одну-скважину делим на цену за барель. Таким образом узнаем минимальное кол-во баррелей, которое необходимое добыть на этих месторождениях во избежание убытка.

```python
product_for_zero_profit = BUDGET_PER_WELL / PRICE
print(f' Minimal stock for well {product_for_zero_profit:.2f} thousand barrels')
```

**Вывод**:
* Если из отобранных месторождений средний запас нефти будет меньше чем 111,11 тыс баррелей - такой регион будет счиаться убыточным.

## Расчёт прибыли и рисков 

```python
names = []
mean_prof = []
loss_perc = []
```

### Функции.

Функция profit позволяет посчитать прибыль для конкретного сэмпла из региона, беря топ-200 лучших скважин в нем. Расчет в млрд. рублей.

```python
def profit(target, predictions):
    pred_sorted = predictions.sort_values(ascending=False)
    selected = target[pred_sorted.index][:TOP_WELLS]
    profit = (selected.sum() * PRICE - BUDGET) / 1_000_000_000
    return profit
```

С помощью функции bootstrap мы можем заданное кол-во раз сымитировать случайный выбор 500 скважин из каждого региона, выбрать из них Топ-200 по предскзааниям, и посчитать для них выручку в млрд. рублей.

По итогам ресемплинга она считает:
* Среднюю выручку для данного региона
* 95% доверительный интервал для этого среднего.
* Нижний порог вероятности убытка в даннмо регионе. в процентах.

```python
def bootstrap(target, predictions, n_iterations, title):
    print(f'Report for {title} region')
    state = np.random.RandomState(RANDOM_STATE)

    profits = []
    for i in range(n_iterations):
        target_subsample = target.sample(n=SEARCH_WELLS, replace=True, random_state=state)
        preds_subsample = predictions[target_subsample.index]    
        profits.append(profit(target_subsample, preds_subsample))

    profits = pd.Series(profits)
    
    sns.histplot(profits).set_title(f'Profit predictions for {title}')
    mean = profits.mean()
    upper = profits.quantile(0.975)
    lower = profits.quantile(0.025)
    loss_percentile = stats.percentileofscore(profits,-1*10**(-9))/100
    
    names.append(title)
    mean_prof.append(mean)
    loss_perc.append(loss_percentile*100)
    
    print(f'Average profit: {mean:.2f} billion rubles')
    print(f'95% confidence interval for profit: [{lower:.2f} billion rubles: {upper:.2f} billion rubles] ')
    print(f'Risk of loss, percentage: {loss_percentile:.2%}')
```

Кол-во итераций по условию зададим равным 1000.

```python
N_ITERATIONS = 1000
```
